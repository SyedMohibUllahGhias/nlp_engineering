```python
import spacy

# 1. Load the model (ensure you ran 'uv run python -m spacy download en_core_web_sm' first)
nlp = spacy.load("en_core_web_sm")

text = "Hello there! Mr. Smith is waiting for you. It's 9:00 A.M. right now."

# 2. Create the Doc object
doc = nlp(text)

# --- Sentence Tokenization ---
print("--- SENTENCES ---")
for i, sent in enumerate(doc.sents, 1):
    print(f"Sentence {i}: {sent.text}")

# --- Word Tokenization ---
print("\n--- WORDS (TOKENS) ---")
# Using a loop for enumerate as requested
for i, token in enumerate(doc, 1):
    print(f"Token {i}: {token.text}")
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
stop_words = set(stopwords.words('english'))
text = "The fox is Hunting for the Dog but he jumped the Bridge."

tokens = word_tokenize(text.lower())

# "filter = [t for t in tokens if t not in stop_words and t.isalpha()]"
# Same below

filtered_tokens = []

for t in tokens:
    if t not in stop_words and t.isalpha():
        filtered_tokens.append(t)
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
```

```python
lemmatizer  = WordNetLemmatizer()
print(lemmatizer.lemmatize("heppiness",pos="n"))
print(lemmatizer.lemmatize("running",pos="n"))
```

```python
from nltk.stem import PorterStemmer

# Initialize the stemmer
stemmer = PorterStemmer()
# Using your specific example words
print(stemmer.stem("heppiness"))
# Output: 'heppiess' (Remains unchanged because it's misspelled)
print(stemmer.stem("running"))
# Output: 'run' (The 'ing' is stripped away)
```

```python
# =========================================================================
# COMMON POS TAGS (Part of Speech)
# Used for NLTK WordNetLemmatizer and SpaCy
# -------------------------------------------------------------------------
# Tag      | Part of Speech | Examples
# ---------|----------------|----------------------------------------------
# 'n'      | Noun           | cat, happiness, London, running (as a sport)
# 'v'      | Verb           | run, jumping, is, sleep
# 'a'/'j'  | Adjective      | happy, quick, blue, better
# 'r'      | Adverb         | quickly, very, happily
# =========================================================================
```
